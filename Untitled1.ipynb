```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = (np.arange(20)).reshape((1,-1))
y = (np.arange(20)*5).reshape((1,-1))
data = np.concatenate((x,y),axis=0)
data
```

```python
plt.plot(x,y)
```

```python
data_mean = (data - np.mean(data,axis = 1).reshape((-1,1))).T
u,s,vh = np.linalg.svd(data_mean)
```

```python
plt.plot(data_mean @ vh[:,0],data_mean @ vh[:,1])
```
